```python
def json_to_csv(data_json, csv_filepath):
    csv_data = open('{}.csv'.format(csv_filepath), 'w', encoding='utf-8')
    csv_writer = csv.writer(csv_data)
    
    for index, entry in enumerate(data_json):
        if index == 0:
            csv_writer.writerow(entry.keys())
        csv_writer.writerow(entry.values())


    csv_data.close()
```

## Heroes Data

```python
OPEN_DOTA_HEROES = 'https://api.opendota.com/api/heroes/?api_key=' + MY_API_HEROES

def get_heroes_response():
    '''Get heroes data request'''
    try:
        response = requests.get(OPEN_DOTA_HEROES)
        data_json = json.loads(response.content.decode('utf-8'))
    except:
        data_json = None
    return data_json
```

```python
json_data = get_heroes_response()
json_to_csv(json_data, 'heroes_data')
```

## Get all Pro player heroes stats

```python
def get_pro_stats(account_id):
    data = get_player_stats_response(account_id)
    json_to_csv(data, './pro_players/{}'.format(str(account_id)))
    
    player = pd.read_csv('./pro_players/{}.csv'.format(str(account_id)))
    df = pd.merge(player, heroes, on='hero_id')
    df = df[['hero_id', 'localized_name', 'games', 'win']]
    
    # Score Algo
    df['score'] = df['win'] * df['win'] / df['games']

    df['normalized_score'] = 100.0 * df['score'] /  df['score'].sum()
    df = df.sort_values(['normalized_score'], ascending=False)
    
    return df[['hero_id', 'normalized_score']]
```

```python
pro_playes = pd.read_csv('pro_players_data.csv')
pro_playes = pro_playes[['account_id', 'personaname']]
pro_playes.head()
```

```python
heroes = pd.read_csv('heroes_data.csv')
heroes = heroes[['id', 'localized_name']]
heroes = heroes.rename(columns={'id': 'hero_id'})
heroes.head()
```

```python
data = {}
cnt = 0
for pro_player in list(pro_playes['account_id'].values):
    try:
        df = get_pro_stats(pro_player)

        heroes_score = {}
        for i, hero in enumerate(list(df['hero_id'].values)):
            heroes_score[str(hero)] = df['normalized_score'][i]
        data[str(pro_player)] = heroes_score
    except:
        continue
    cnt += 1
    if cnt % 5 == 0:
        print(str(cnt) + ' pro player data collected')
```

```python
with open('data.json', 'w') as outfile:
    json.dump(data, outfile)
```
